##**Convolutional Neural Network (CNN):**

**Strengths**

**Feature extraction**

CNN is good at automatic feature extraction from images which enhance them to learn more complex samples. For example, this can be used in healthcare to make them learn complex samples relevant to medical diagnoses.

**Hierarchical representation**

One of the strengths of CNN is hierarchical representation this refers to the idea that the network learns to recognize features at different stages of complexity, in order to represent this in CNN is by using multiple convolutional layers.

This starts from the early layer which their main purpose is to learn small and simple features in the input data like edges, corners, and textures but as it moves to the next layers, the next layer starts to learn from the previous layer to combine these features to convert them into complex patterns or shapes, until it reaches the point where the network can recognise large structures or objects. This is really useful in order to make accurate predictions in image classification for healthcare purposes.


**Weakness**

**Interpretability**

On another hand interpretability is one of CNN weakness, this is referred to the ability to understand and explain how the model makes decisions. As we have mention before CNN uses hierarchical representation which makes the network complex as it uses features that after it will get combine and converted into complex patterns, this makes it challenging for humans to understand how each features get involved and affect to the final decision.  

**Data bias**

Another weakness in CNN is data bias this means that the data may lead to biased predictions. In the neural network there are layers and each layer have bias which is an additional parameter. Biases have disadvantages which are overfitting, and underfitting. For example, too much bias can end up in overfitting the model in the training data so it will perform bad in the new data. Whereas underfitting is using few biases which means bad performance in the training set.


**Advantages**

**High accuracy**

One of the advantages of CNN is the ability to get high accuracy rates, this is because the network learns to recognise features at different levels of complexity and also depends on the dataset, the larger the dataset the better as model will have more data input to train and this will help the model learn to recognise specific features with great accuracy.

**Automation**

 Automated images classification using CNN reduce the dependence on human expertise which means they can learn to recognize patterns in images with no human intervention.  Automation in CNN is ideal as the model has the ability to learn both feature extraction and task specific predictions which means better performance.


**Disadvantages**

**High computational requirements**

One of the disadvantages of using CNN is their high computational requirements which are expensive, requires powerful hardware, this is because CNN tend to have large number of layers and parameters therefore it uses processing power and memory in order to train and run.

**Small dataset**

Another disadvantage thar CNN has is that it requires large datasets in order to get high accuracy rates. As I have mentioned the model uses this dataset to learn to recognise patterns based on the dataset images, so if the dataset is small the CNN will overfit causing a bad performance with a new data.

**Type of input**

The type of input data in CNN in this case is for image classification which the  input data can be MRI, CT scans and X-rays for healthcare.

**Expected outcome**

##**Transfer Learning:**

**Strengths**

**Improved performance**

One of the strengths of transfer learning is improved performance, this is because of its useful pre-training. Pre-training helps with faster training times as there is not needed to start the model from scratch. So, transfer learning is like transferring the knowledge from the pre-training model as that model has already learnt that task and apply it to a new related task in order to enhance the performance.

**Data efficiency**

Data efficiency is another strength in transfer learning this is because in some cases trying to get large label dataset gets difficult. Transfer learning is a technique from machine learning and machine learning uses label or marked data in order to instruct the model with information that has to learn. However, by using transfer learning we do not have to provide as much labelled data to the specific task.


**Weakness**

**Domain mismatch**

One of the weaknesses in transfer learning is domain mismatch, this occurs when you use a trained model with one data content to use it on a target with totally different data content and this affect the performance that leads to a poor performance. in that case transfer learning is not that better than training model from scratch. For example, you use a pre-trained model with a dataset source from ImageNet in order to use it for image classification cancer disease dataset, these two datasets are different domains, and we use transfer learning when there is a relation between pre-trained model and the target dataset.

**Overfitting**

Another weakness of transfer learning is overfitting, this can occur when you have a small dataset, but your pre-trained model has a large dataset, and you try to fine-tuning on a small dataset this can lead to overfitting. This is because transfer learning has larger dataset which cause to learn the training data too well so when this is applied into the new task it can overfit the initial data and might not generalize well in the new task.

**Advantages**

**Training time efficiency**

One of the advantages in transfer learning is training time efficiency, this is because transfer learning uses pre-trained model that has learn to recognise generic features in a large dataset, so rather than build a network from scratch which is expensive and time consuming, transfer learning saves time because pre-trained model is has already learn to recognise generic features which can be use to apply into a related new task.

**Small dataset**

Another advantage of transfer learning is that even if you do not have a large dataset for training data in order to train your network, you can use a pre-trained model as this pre-trained model has already been trained with a large amount of dataset which provides you the knowledge of a pre-trained model on a large dataset, this means that it will give you a better generalization and be more effective in cases when you have small dataset. Transfer learning only works if the pre-trained model is related to the new task.


**Disadvantages**

**Limited customization**

One of the disadvantages of transfer learning is limited customization. As transfer learning provides knowledge that the pre-trained model has already learnt to a related new task issues may arise in some cases when it comes to fine-tuning, as you may not be able to fine-tuning the learned knowledge when you apply to a new task this means that as the model has limited customization which will be difficult for the model to adapt to the new problem task.

**Interpretability**

Another disadvantage of transfer learning is interpretability. In transfer learning the pre-trained model is fine-tuning in order to use it in a new task, the problem arise when the process adaptation into the new task is not able to interpret it. This becomes challenging in order to understand how the knowledge of the pre-trained model is used into the new task, because you do not understand how the features in the pre-trained model are affecting the decision-making process in the new task.

**Type of input**

Transfer learning uses two types of input which are the dataset used in the pre-trained model that can use ImageNet for image classification and the dataset used in the new specific problem task can be MRI, CT scans and X-rays for healthcare.

**Expected outcome**

The expected outcome depends on the specific task. For example, I can use transfer learning to classify images about brain tumour based on binary classification like Tumor or Normal which the evaluation will depend on the accuracy rate to predict these images.

##**Support vector machines (SVMs):**

**Strengths**

**Robust to overfitting**

 One of the strengths of SVMs is robust overfiring. SVMs are less likely to overfitting which is a good point when you are dealing with limited label datasets. This is achieved in SVMs because SVMs focus on finding the hyperplane that maximises the margin between different classes which is the distance between line or hyperplane and the nearest data point of each class. So, as SVMs focus on the margin maximization this leads to select a decision boundary and avoid noise in the training data which generalizes good with new data and prevents overfitting.

**Good at binary classification**

Another strength in SVMs is binary classification. This is because the main purpose of support vector machines is binary classification. Classifying it into two classes like normal or cancer, this makes SVMs so much easier and efficient when it comes to classify two-class problems. So, when SVMs finds the hyperplane that separates the data into to classes in a binary classification scenario, it helps to find the hyperplane between the two classes and make the decision boundary.

**Weakness**

**Limited scalability**

One of the weaknesses in SVMs is limited scalability. Limited scalability means how well it performs when the size of the dataset increases. This is because SVMs uses complex mathematical calculations, so the larger the dataset and features the more complex mathematical calculations SVMs will be using. Also, the training time will be involved as the process can be time consuming which makes the SVMs slower specially when a large dataset is being used.

**Large scale image data**

Another weakness of SVMs is large scale image data. SVMs are good at image classification however, SVMs have a limitation when is about large scale image data because in high dimensional data each pixel is equivalent to one feature. For example, in 100 x 100 pixel image there is 10,000 features, so if the scale of the image increases the features will increase too meaning that the complex mathematical calculations will increase too which leads to time consuming in order to find the hyperplane.  

**Advantages**

**High dimensionality**

One of the advantages in SVMs is high dimensionality. SVMS works with high dimensional spaces which helps to find complex decisions boundaries rather than use lower dimensions which may not be achievable. The aim of SVMs is to find the best hyperplane that separates the data points that belong to different classes, and this can be achieved by using high dimensional space where there can be more possible ways in order to find the best hyperplane.

**Regularization**

Another advantage of SVMs is regularization. Regularization can used when SVMs try to find the hyperplane that separates different classes as a way of prevention from becoming too complex leading to overfitting the training data with noise.

**Disadvantages**

**Sensitivity noisy data**

One of the disadvantages of SVMs is sensitivity noisy data. SVMs can be affected by random unpredictable fluctuations in the training data set which that noise does not let it identify the target patterns properly. So, this can lead to overfitting because the model has learned the training data that performs very well however when new a new dataset comes it will give a poor performance.

**Interpretability**

Another disadvantage of SVMs is interpretability. The reason is that the goal of SVMs is to find the hyperplane in order to separate classes, and the hyperplane is determined by the data points that nearest to the hyperplane. However, to understand and explain the decision boundary gets difficult as SVMs can be very complex and non-linear due to the use of high dimensional spaces.

**Type of input**

The type of input data in support vector machines (SVMs) in this case is for image classification which can be MRI, CT scans and X-rays for healthcare.


**Expected outcome**

The expected outcome depends on the specific task. For example, I can use support vector machines (SVMs) to classify images about brain tumour based on binary classification like Tumor or Normal which the evaluation will depend on the probability scores for the classification.

```python
import os
import cv2
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from glob import glob
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Flatten, Input, MaxPooling2D, Dropout
from sklearn.metrics import classification_report
```

### **Preprocessing**

```python
img_size = 200
batch_size = 32

traingen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1/255.0,
                                                           #brightness_range = [0.2,1.2],
                                                           height_shift_range = 0.1,
                                                           shear_range = 0.2,
                                                           zoom_range = 0.2,
                                                           horizontal_flip = True)

valdgen = tf.keras.preprocessing.image.ImageDataGenerator(height_shift_range = 0.1,
                                                           horizontal_flip = True)

testgen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1/255.0)

#/content/drive/MyDrive/Brain_Tumor_Dataset/training_set
traindata = traingen.flow_from_directory('/content/drive/MyDrive/Brain_Dataset/train_set',
                                         target_size = (img_size, img_size),
                                         batch_size = batch_size,
                                         shuffle = True,
                                         class_mode = 'binary',
                                         color_mode = 'grayscale')

valid_Data = valdgen.flow_from_directory('/content/drive/MyDrive/Brain_Dataset/valid_set',
                                         shuffle=False,
                                         batch_size=batch_size,
                                         target_size=(img_size,img_size),
                                         class_mode='binary',
                                         color_mode='grayscale')

testdata=testgen.flow_from_directory('/content/drive/MyDrive/Brain_Dataset/test_set',
                                     shuffle=False,
                                     batch_size=batch_size,
                                     target_size=(img_size,img_size),
                                     class_mode='binary',
                                     color_mode='grayscale')
```

### **Images for each class in each Dataset**


1.   Training set
2.   Test set
3.   Validation set



```python
df_TrnD = pd.DataFrame(traindata.classes)

df_TstD = pd.DataFrame(testdata.classes)

df_VD = pd.DataFrame(valid_Data.classes)


print("-"*60)
print("Number of Images in Train Data in each class")
print(df_TrnD.value_counts())
print("-"*60)
print("Number of Images in Test Data in each class")
print(df_TstD.value_counts())
print("-"*60)
print("Number of Images in Validation Data in each class")
print(df_VD.value_counts())
print("-"*60)
```

### **Displaying some images from the training dataset**

```python
#Clases normal or tumor
labels = ['NORMAL', 'TUMOR']

samples=traindata.__next__()

images = samples[0]
target = samples[1]

for i in range(20):
  plt.subplot(5,4, i+1)
  plt.subplots_adjust(hspace = 1.0)
  plt.imshow(images[i])
  plt.title(f" Label: {labels[int(target[i])]} ")
  plt.axis('off')
```

### **Build Neural Network Model**

```python
model = Sequential()

#Block 1
model.add(Conv2D(32, (3, 3), strides=1, activation='relu', input_shape=(img_size, img_size, 1), padding='same'))
model.add(Conv2D(32, (3, 3),strides=1, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2,2), strides=2))

#Block 2
model.add(Conv2D(64, (3, 3), strides=2, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2,2), strides=2))

#Block 3
model.add(Conv2D(128, (3, 3), strides=1, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2,2), strides=2))
model.add(Dropout(0.2))

#Top Layers
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

### **Compiling the model**

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

### **Train the model**

```python
epochs = 15
history = model.fit(traindata, validation_data = valid_Data, epochs = epochs,)
```

### **Evaluate the model**

```python
train_loss, train_acc = model.evaluate(traindata, verbose=2)

print('\nTrain accuracy:', train_acc)
print('\nTrain loss:', train_loss)
```

```python
vald_loss, vald_acc = model.evaluate(valid_Data, verbose=2)

print('\nValidation accuracy:', vald_acc)
print('\nValidation loss:', vald_loss)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

##**Testing the model**

### **Get Probabilities**

```python
ytest = testdata.classes
probs = model.predict(testdata)
print('Size of variable probabilities: ' + str(probs.shape)) # You can see the size of the arrays here (300, 1)
print('Probabilities:')
print(probs) #Returns an array of 300 arrays with one probability as it is a binary classification
for i in range(len(labels)):
    print(labels[i] + ': ' + str(np.count_nonzero(ytest == i)))
```

```python
probs[probs > 0.5] = 1
probs[probs < 0.5] = 0
print('Predicted image as label:')
print(probs) #Displays a converted array to 0 or 1, if the probability is greater than 0.5 will be 1 if it is less than 0.5 will be 0
print('Actual image label:')
print(ytest) #Displays array of actual image labels
```

### **Precision, Recall and F1 Score**

```python
print(classification_report(ytest, probs, target_names= labels))
```

Based on the Precision, Recall and F1 Score we can see that the model has a good precision on the tumor class than the normal class. However, the recall is better in normal class than in tumor class. As we can see the accuracy is 0.90 which shows the performance of the model at predicting is good.

### **Confusion Matrix**

```python
def plot_confusion_matrix(actual, predicted, labels):
  cm = tf.math.confusion_matrix(actual, predicted)
  ax = sns.heatmap(cm, annot=True, fmt='g')
  sns.set(rc={'figure.figsize':(7, 7)})
  ax.set_title('Confusion matrix')
  ax.set_xlabel('Prediction')
  ax.set_ylabel('Label')
  plt.xticks(rotation=90)
  plt.yticks(rotation=0)
  ax.xaxis.set_ticklabels(labels)
  ax.yaxis.set_ticklabels(labels)

plot_confusion_matrix(ytest, probs, labels)
```

As we can see on the confusion matrix there was 150 images for class normal and 150 images for class tumor, the expectation is to get all the 150 tumor images get predicted as tumor and all the 150 normal images get predicted as normal.

However, it has predicted 142 images as normal and in fact these images belong the normal class but only 8 images where predicted as tumor.

On the another hand 127 images where predicted as tumor and they were tumor, but only 23 images where predicted as normal. This shows the accuracy at predicting these images is good.

```python
#this function will display a message for the predicted label and the image
def predict_img(probs, img, labels):
  if probs > 0.5:
    print(f'This image has been predicted as Label: {labels[1]}')
    print(f'Predicted Value: {probs}')
    plt.imshow(img)
    plt.title(labels[1])
    plt.show()
  else:
    print(f'This image has been predicted as Label: {labels[0]}')
    print(f'Predicted Value: {probs}')
    plt.imshow(img)
    plt.title(labels[0])
    plt.show()
```
